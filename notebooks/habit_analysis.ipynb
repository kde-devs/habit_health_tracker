```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

path = '/content/drive/MyDrive/Colab Notebooks/data/daily_habits.csv'
df = pd.read_csv(path)
df['date'] = pd.to_datetime(df['date'])

plt.figure(figsize = (12, 6))
for col in ['exercise', 'study', 'health', 'self_care', 'diet']:
  plt.plot(df['date'], df[col], label = col)

plt.title('일일 건강 습관 점수 (2025년 7월)')
plt.xlabel('날짜')
plt.ylabel('점수 (0~5)')

plt.legend(loc = 'center left', bbox_to_anchor = (1.02, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```
